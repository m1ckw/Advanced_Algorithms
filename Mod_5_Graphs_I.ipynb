```python
class Node:
    
    def __init__(self, v):

        self.value = v         
        self.inNeighbors = []
        self.outNeighbors = []
        self.status = "unvisited"
        self.colour = 0 # Added for the bipartitie check
        
    def hasOutNeighbor(self, v):
        
        if v in self.outNeighbors:
            return True
        return False
        
    def hasInNeighbor(self, v):
        if v in self.inNeighbors:
            return True
        return False
    
    def hasNeighbor(self, v):
        if v in self.inNeighbors or v in self.outNeighbors:
            return True
        return False
    
    def getOutNeighbors(self):
        return self.outNeighbors
    
    def getInNeighbors(self):
        return self.inNeighbors
        
    def addOutNeighbor(self, v):
        self.outNeighbors.append(v)
    
    def addInNeighbor(self, v):
        self.inNeighbors.append(v)
    
    def __str__(self):
        return str(self.value) 
```

```python
class Graph:
    
    def __init__(self):
        
        self.vertices = []

    def addVertex(self,n):
        
        self.vertices.append(n)
        
    # add a directed edge from u to v
    def addDiEdge(self, u, v):
        
        u.addOutNeighbor(v)
        v.addInNeighbor(u)
        
    # add edges in both directions between u and v
    def addBiEdge(self, u, v):
        
        self.addDiEdge(u, v)
        self.addDiEdge(v, u)
        
    # get a list of all the directed edges
    # directed edges are a list of two vertices
    def getDirEdges(self):
        
        ret = []
        for v in self.vertices:
            ret += [ [v, u] for u in v.outNeighbors ]
        return ret
    
    def __str__(self):
        ret = "Graph with:\n"
        ret += "\t Vertices:\n\t"
        for v in self.vertices:
            ret += str(v) + ","
        ret += "\n"
        ret += "\t Edges:\n\t"
        for a,b in self.getDirEdges():
            ret += "(" + str(a) + "," + str(b) + ") "
        ret += "\n"
        return ret
```

```python
# Post-Order DFS - Code Provided
def DFS_PostOrder(w): 
    # Set the nodes status to 'inprogress'
    w.status = "inprogress"
    
    # Check for out bound vertices
    for v in w.getOutNeighbors(): 
        # if outbond vertex unvisted, pass it to DFS helper
        if v.status == "unvisited":   
            DFS_PostOrder(v)
    # Print the value of the current Vertex
    print(w)
    # Sets the status to done
    w.status = "done"
    
        
def DFS_Post(w, G):
    # Sets all vertices as "unvisted".
    for v in G.vertices:
        v.status = "unvisited"
    # Once all nodes have been set as unvisted, calls the helper.     
    return DFS_PostOrder(w)
```

```python
# Created a copy of the above code and made a few slight adjustments 
# to obtain pre-order form output. 

def DFS_PreOrder(w): 
    # Set the nodes status to 'inprogress'
    w.status = "inprogress"
    # Print the value of the current Vertex
    print(w)
    
    # Check for out bound vertices
    for v in w.getOutNeighbors(): 
        # if outbond vertex unvisted, pass it to DFS helper
        if v.status == "unvisited":   
            DFS_PreOrder(v)
    
    # Sets the status to done
    w.status = "done"
    
        
def DFS_Pre(w, G):
    # Sets all vertices as "unvisted".
    for v in G.vertices:
        v.status = "unvisited"
    # Once all nodes have been set as unvisted, calls the helper.     
    return DFS_PreOrder(w)
```

Like the DFS algorithm BFS also starts by setting all the vertex's statuses to “unvisited” before taking the length of the vertex list and storing as n. That variable is than used to create an array of empty arrays (Ls), one for each node. The number of empty arrays is set equal to the number of nodes to capture the deepest possible graph i.e., in case each node is a child of the node before it.

Then, the fist empty array is assigned the root node passed to the BFS(w, G) function. Once set, the algorithm moves into the outer for loop (for_i) and then immediately into the next loop, for_u which sets the variable u to the value of Ls[0] the root node. The algorithm prints out the value of the node before moving into the final loop, for_v.  Here the variable v is set equal to the out bound array contained within the current node. Then moves through the unvisited nodes appending them to the next array position current node +1. These loops work their way through the vertices of the graph storing the nodes on their relevant level, root = 0, children of root = 1, so on.

### Section 7: Developing Algorithm for Strongly Connected Components

We will have to modify our DFS to incorporate the in and out time. We will have to modify our graph as well

Let us start by adding inTime and outTime in the Node first

```python
class Node2:
    
    def __init__(self, v):

        self.value = v
        self.inNeighbors = []
        self.outNeighbors = []
        
        self.status = "unvisited"
        self.inTime = None
        self.outTime = None
        
    def hasOutNeighbor(self, v):
        
        if v in self.outNeighbors:
            return True
        return False
        
    def hasInNeighbor(self, v):
        
        if v in self.inNeighbors:
            return True
        return False
    
    def hasNeighbor(self, v):
        if v in self.inNeighbors or v in self.outNeighbors:
            return True
        return False
    
    def getOutNeighbors(self):
        return self.outNeighbors
    
    def getInNeighbors(self):
        return self.inNeighbors
        
    def addOutNeighbor(self, v):
        self.outNeighbors.append(v)
    
    def addInNeighbor(self, v):
        self.inNeighbors.append(v)
    
    def __str__(self):
        return str(self.value) 
```

We should modify our Graph data structure to be able to reverse edges in the graph, we will do this by adding an extra function reverseEdge(u, v)

```python
class Graph2:
    
    def __init__(self):
        
        self.vertices = []

    def addVertex(self,n):
        
        self.vertices.append(n)
        
    # add a directed edge from CS161Node u to CS161Node v
    def addDiEdge(self, u, v):
        
        u.addOutNeighbor(v)
        v.addInNeighbor(u)
        
    # add edges in both directions between u and v
    def addBiEdge(self, u, v):
        
        self.addDiEdge(u, v)
        self.addDiEdge(v, u)
        
    # get a list of all the directed edges
    # directed edges are a list of two vertices
    def getDirEdges(self):
        
        ret = []
        for v in self.vertices:
            ret += [ [v, u] for u in v.outNeighbors ]
        return ret
    
    # reverse the edge between u and v.  Multiple edges are not supported.
    def reverseEdge(self,u,v):
        
        if u.hasOutNeighbor(v) and v.hasInNeighbor(u):
            
            if v.hasOutNeighbor(u) and u.hasInNeighbor(v): 
                return
        
            self.addDiEdge(v, u)
            u.outNeighbors.remove(v)
            v.inNeighbors.remove(u)        
                
    def __str__(self):
        ret = "Graph with:\n"
        ret += "\t Vertices:\n\t"
        for v in self.vertices:
            ret += str(v) + ","
        ret += "\n"
        ret += "\t Edges:\n\t"
        for a,b in self.getDirEdges():
            ret += "(" + str(a) + "," + str(b) + ") "
        ret += "\n"
        return ret
```

Okay, let us now modify our DFS to include in and out time during graph traversal

```python
def DFS2(w, currentTime, ordering, verbose):
    
    if verbose:
        print("Time", currentTime, ":\t entering", w)
    
    w.inTime = currentTime
    currentTime += 1
    w.status = "inprogress"
    
    for v in w.getOutNeighbors():
        
        if v.status == "unvisited":
            currentTime = DFS2(v, currentTime, ordering, verbose)
            currentTime += 1
    
    w.outTime = currentTime
    w.status = "done"
    ordering.insert(0, w)
    
    if verbose:
        print("Time", currentTime, ":\t leaving", w)
    
    return currentTime
```

An now the SCC algorithm:

```python
def SCC(G, verbose=False):
    
    ordering = []
    
    for v in G.vertices:
        v.status = "unvisited"
        v.inTime = None
        v.outTime = None
    
    currentTime = 0
    for w in G.vertices:
        
        if w.status == "unvisited":
            currentTime = DFS2(w, currentTime, ordering, verbose)
        currentTime += 1
    
    # now reverse all the edges
    E = G.getDirEdges()
    for x,y in E:
        G.reverseEdge(x,y)
   
    # and do it again, but this tim in the order "ordering"
    SCCs = []
    for v in ordering:
        v.status = "unvisited"
        v.inTime = None
        v.outTime = None
    
    currentTime = 0
    for w in ordering:
        visited = [] 
        
        if w.status == "unvisited":
            currentTime = DFS2(w, currentTime, visited, verbose)
            SCCs.append(visited[:])
    
    return SCCs
```

### <font color='red'> Activity 5 - Code Reflections. </font>

The SCC algorithm uses the DFS with a couple of small changes. Firstly, the in and out times have been added as attributes to the node in the Node class, and secondly, a reverseEdge() function was added to the Graph class. As per the earlier functions, the first step in the SCC algorithm is to ensure all the node values have been reset, that is ‘unvisited’ and no in or out times.   

Then starting from the root, if the node is “unvisited”, the updated DFS function is called passing the current node, current time, the “ordering” array and the boolean value for verbose. This is a standard DFS recursive call that works its way to the leaf nodes setting the status of each node to “in progress” and stamping the time of each node in order of being visited along the way. 

Once at the bottom of the graph, the leaf node is the first to be added to the ordering array. When leaving the leaf node to traverse back to the top (the root), the out time is also stamped, and the status is set to “done”. Each node on the way back to the root is added to position [0] of the order array creating a top to bottom order. 

Once at the root, the getDirEdges() method is called and the graph is traversed again collecting the direction of each edge. Once all the directed edges have been collected and returned, the new reverseEdge() function is called to reverse those edges that are not already connected in both directions. 

Once the edges have been reversed, the SCCs array is instantiated as an empty array and the process of resetting each node starts again. The current time variable is set to zero, a new array called visited is created, and a new DFS is called on the ordering array traversing from the same root node.  

This time through, the nodes are added to the SCC’s array in order of their visit. If the elements traversed in the ordering array, are visited (in reverse) from the current node, they are added to the “visited array”. Once the node's outbound neighbours have been explored, the node’s status is marked as done and the visited array is added to the SCCs array. The result of which is arrays of strongly connected elements. Otherwise, the array contains only a single node.
